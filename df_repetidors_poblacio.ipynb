```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import unicodedata
import os
from unidecode import unidecode
```

```python
df_repetidors_poblacio_2022 = pd.read_csv('clean_data/2022/df_repetidors_poblacio.csv', delimiter=',')
df_repetidors_poblacio_2021 = pd.read_csv('clean_data/2021/df_repetidors_poblacio.csv', delimiter=',') 
```

```python
datasets = {
    'df_repetidors_poblacio_2022': df_repetidors_poblacio_2022,
    'df_repetidors_poblacio_2021': df_repetidors_poblacio_2021,
}

# Iterar sobre el diccionari per mostrar les primeres files de cada dataset
for name, df in datasets.items():
    print(f"Info Dataset {name}:")
    display(df.info()) 
    print("\n") 
```

```python
# Crear una llista de DataFrames amb els seus anys corresponents
datasets = [
    (df_repetidors_poblacio_2022, 2022),
    (df_repetidors_poblacio_2021, 2021),
]

# Afegir la columna 'any' i concatenar tots els DataFrames
df_repetidors_poblacio = pd.concat([df.assign(any=year) for df, year in datasets], ignore_index=True)

# Mostrar la informació final
df_repetidors_poblacio
```

```python
# Convertir la columna 'codi' a tipus string
df_repetidors_poblacio['codi'] = df_repetidors_poblacio['codi'].apply(lambda x: str(x).zfill(2))
display(df_repetidors_poblacio.head())
df_repetidors_poblacio.info()
```

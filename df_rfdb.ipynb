```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import unicodedata
import os
from unidecode import unidecode
```

```python
df_rfbd_2021 = pd.read_csv('clean_data/2021/df_rfbd.csv', delimiter=',') 
df_rfbd_2020 = pd.read_csv('clean_data/2020/df_rfbd.csv', delimiter=',') 
df_rfbd_2019 = pd.read_csv('clean_data/2019/df_rfbd.csv', delimiter=',') 
df_rfbd_2018 = pd.read_csv('clean_data/2018/df_rfbd.csv', delimiter=',')
```

```python
datasets = {
    'df_rfbd_2021': df_rfbd_2021,
    'df_rfbd_2020': df_rfbd_2020,
    'df_rfbd_2019': df_rfbd_2019,
    'df_rfbd_2018': df_rfbd_2018,
}

# Iterar sobre el diccionari per mostrar les primeres files de cada dataset
for name, df in datasets.items():
    print(f"Info Dataset {name}:")
    display(df.info()) 
    print("\n") 
```

```python
# Crear una llista de DataFrames amb els seus anys corresponents
datasets = [
    (df_rfbd_2021, 2021),
    (df_rfbd_2020, 2020),
    (df_rfbd_2019, 2019),
    (df_rfbd_2018, 2018)
]

# Afegir la columna 'any' i concatenar tots els DataFrames
df_rfbd = pd.concat([df.assign(any=year) for df, year in datasets], ignore_index=True)

# Mostrar la informació final
df_rfbd
```

```python
# Convertir la columna 'codi' a tipus string
df_rfbd['codi'] = df_rfbd['codi'].apply(lambda x: str(x).zfill(2))
display(df_rfbd.head())
df_rfbd.info()
```

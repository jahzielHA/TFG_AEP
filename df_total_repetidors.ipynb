```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import unicodedata
import os
from unidecode import unidecode
```

```python
df_total_repetidors_2022 = pd.read_csv('clean_data/2022/df_total_repetidors.csv', delimiter=',')
df_total_repetidors_2021 = pd.read_csv('clean_data/2021/df_total_repetidors.csv', delimiter=',') 
```

```python
datasets = {
    'df_total_repetidors_2022': df_total_repetidors_2022,
    'df_total_repetidors_2021': df_total_repetidors_2021,
}

# Iterar sobre el diccionari per mostrar les primeres files de cada dataset
for name, df in datasets.items():
    print(f"Info Dataset {name}:")
    display(df.info()) 
    print("\n") 
```

```python
# Crear una llista de DataFrames amb els seus anys corresponents
datasets = [
    (df_total_repetidors_2022, 2022),
    (df_total_repetidors_2021, 2021),
]

# Afegir la columna 'any' i concatenar tots els DataFrames
df_total_repetidors = pd.concat([df.assign(any=year) for df, year in datasets], ignore_index=True)

# Mostrar la informació final
df_total_repetidors
```

```python
# Convertir la columna 'codi' a tipus string
df_total_repetidors['codi'] = df_total_repetidors['codi'].apply(lambda x: str(x).zfill(2))
display(df_total_repetidors.head())
df_total_repetidors.info()
```
